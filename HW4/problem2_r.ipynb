3. Making a new variable saturated_fat_per_gram by dividing fat by weight

```r
# make a new column divinding fat by weight
nutrition_subset$saturated_fat_per_gram <- with(nutrition_subset, saturated_fat/weight_in_grams)

# sort and print out
head(nutrition_subset[order(nutrition_subset$saturated_fat_per_gram, decreasing=TRUE), 1:5],5)
```

4. Standardizing saturated_fat_per_gram and finding outliers.

```r
# histogram
hist(nutrition_subset$saturated_fat_per_gram, col="green")

# make a new column standardizing the column
nutrition_subset$scaled_saturated_fat_per_gram <- scale(x = nutrition_subset$saturated_fat_per_gram)

# summarize
summary(nutrition_subset$scaled_saturated_fat_per_gram)

#three-sigma rule
three_sigma_rule <- (nutrition_subset$scaled_saturated_fat_per_gram < -3 |
                     nutrition_subset$scaled_saturated_fat_per_gram > 3 )
three_sigma_low_end <- (nutrition_subset$scaled_saturated_fat_per_gram < -3)
three_sigma_high_end <- (nutrition_subset$scaled_saturated_fat_per_gram > 3)


# get outliers

cat ("There are", sum(three_sigma_rule), "outliers \n\n")
cat (sum(three_sigma_high_end), "of them are high-end outliers. \n\n")
cat (sum(three_sigma_low_end), "of them are low-end outliers. \n\n")

head(nutrition_subset[order(nutrition_subset$scaled_saturated_fat_per_gram, decreasing=TRUE), 1:6], sum(three_sigma_rule))

# density plots
d1 <- density(nutrition_subset$saturated_fat_per_gram)
plot(d1, main="Kernel Density of Unscaled Saturated Fat per gram")
polygon(d1, col="orange", border="blue")
d2 <- density(nutrition_subset$scaled_saturated_fat_per_gram)
plot(d2, main="Kernel Density of Scaled Saturated Fat per gram")
polygon(d2, col="orange", border="blue")

```

There aren't any outliers at the low end of the scale
