1. Importing the data and converting it.

```r
## import the data

library(readr)
adult_training <- read_csv("adult_training.csv", show_col_types = FALSE)
```

2. Creating and plotting contingency table for workclass vs. sex

```r

## make a contingency table for workclass vs. sex

ctable1 <- table(adult_training$workclass, adult_training$sex)
table(adult_training$workclass, adult_training$sex)

## table proportion
prop.table(ctable1)

## bar plot

barplot(ctable1,
        xlab=names(adult_training$sex),
        las=1, col=rainbow(10))


```

3. Sex vs. martial status

```r
## make a contingency table for sex vs. martial status

ctable2 <- table(adult_training$sex, adult_training$`marital-status`)
table(adult_training$sex, adult_training$`marital-status`)

## table proportion
prop.table(ctable2)

## bar plot

barplot(ctable2,
        xlab=names(adult_training$sex), ylab=names(adult_training$sex), col=rainbow(10))

```

4. Make a dataframe with people over 40 years of age.

5. Recreate the contingency table for sex vs. martial status with the new dataframe.

6. Identify outliers in the education field

```r
# summarize data
summary(adult_training$education)

# histogram
hist(adult_training$education)
```

```r
# scale
adult_training$scaled_education <- scale(x = adult_training$education)
summary(adult_training$scaled_education)

# three sigma rule
three_sigma <- (adult_training$scaled_education < -3 |
                     adult_training$scaled_education > 3 )

# print number of outliers
cat ("There are", sum(three_sigma), "outliers \n\n")

# print the outliers
cat("The scaled age values that are outliers: \n\n")
adult_training$scaled_education[which(three_sigma)]

# print the non-outliers
cat("The scaled age values that aren't outliers: \n\n")
adult_training$scaled_education[which(!three_sigma)]

```

There are indeed outliers in the education field.

7. Standarizing the age variable and identifying the number of outliers as well as the most extreme outlier.

```r
#summarize data
summary(adult_training$age)
hist(adult_training$age)
```

```r
# scale and summarize
adult_training$scaled_age <- scale(x = adult_training$age)
summary(adult_training$scaled_age)

# three sigma rule
three_sigma2 <- (adult_training$scaled_age< -3 |
                     adult_training$scaled_age > 3 )

# print number of outliers
cat ("There are", sum(three_sigma2), "outliers \n\n")

# print the outliers
cat("The scaled age values that are outliers: \n\n")
adult_training$scaled_age[which(three_sigma2)]

# print the non-outliers
cat("The scaled age values that aren't outliers: \n\n")
adult_training$scaled_age[which(!three_sigma2)]
```

There are 60 outliers with the most extreme outlier being 3.7512.
